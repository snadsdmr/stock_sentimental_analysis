```python
import numpy as np
import pandas as pd
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

# Select start date for correlation window as well as list of tickers
start = datetime(2017, 1, 1)

# Define sectors and their respective stocks
sectors = {
    "Technology": ['GOOGL', 'AMZN', 'META', 'NFLX', 'AAPL', 'MSFT', 'BABA', 'ADBE', 'TSLA'],
    "Consumer Discretionary": ['NKE', 'COST', 'SBUX', 'HD', 'TGT', 'PG', 'KO', 'PEP', 'UNH'],
    "Health Care": ['JNJ', 'ABBV', 'BMY', 'LLY', 'REGN', 'AMGN', 'GILD'],
    "Industrials": ['UTX', 'HON', 'BA', 'CAT', 'MMM', 'DD', 'DOW'],
    "Financials": ['V', 'MA', 'AXP', 'WFC', 'BAC', 'PLD'],
    "Utilities": ['NEE', 'AEP', 'D', 'DUK', 'EXC']
}

# Initialize an empty dictionary to store correlation data for each sector
sector_correlations = {}

# Fetch historical stock prices and calculate correlations for each sector
for sector, symbols_list in sectors.items():
    symbols = []

    # Fetch historical stock prices using yfinance for each symbol in the sector
    for ticker in symbols_list:
        data = yf.download(ticker, start=start)
        data['Symbol'] = ticker
        symbols.append(data)

    # Concatenate the dataframes into a single dataframe for the sector
    df = pd.concat(symbols)

    # Reset the index and select relevant columns
    df = df.reset_index()
    df = df[['Date', 'Close', 'Symbol']]

    # Pivot the dataframe
    df_pivot = df.pivot('Date', 'Symbol', 'Close').reset_index()

    # Calculate the correlation matrix for the sector
    corr_df = df_pivot.corr(method='pearson')

    # Store the correlation matrix in the sector_correlations dictionary
    sector_correlations[sector] = corr_df

# Plot a heatmap of the correlations for each sector
for sector, corr_df in sector_correlations.items():
    plt.figure(figsize=(10, 8))
    mask = np.zeros_like(corr_df)
    mask[np.triu_indices_from(mask)] = True
    sns.heatmap(corr_df, cmap='RdYlGn', vmax=1.0, vmin=-1.0, mask=mask, linewidths=2.5, annot=True, fmt=".2f")
    plt.title(f'Correlations for {sector} Sector')
    plt.yticks(rotation=0)
    plt.xticks(rotation=90)
    plt.show()
```
